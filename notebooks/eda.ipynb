```python
# Boxplot to detect outliers
sns.boxplot(data['TotalCharges'])
plt.show()

# Z-score method to detect outliers
from scipy import stats
z_scores = stats.zscore(data.select_dtypes(include=['float64', 'int64']))
abs_z_scores = abs(z_scores)
outliers = (abs_z_scores > 3).all(axis=1)
print(data[outliers])
```

```python
# Load the dataset
data = pd.read_csv("processed_data.csv")  # Replace with the path to your dataset
print(data.head())  # Verify dataset is loaded
```

```python
bins = [0, 12, 24, 36, 48, 60, data['tenure'].max()]
labels = ['0-1 Year', '1-2 Years', '2-3 Years', '3-4 Years', '4-5 Years', '5+ Years']

# Create a new column for Tenure Groups
data['TenureGroup'] = pd.cut(data['tenure'], bins=bins, labels=labels)

# Verify the grouping
print(data[['tenure', 'TenureGroup']].head())
```

```python
# Calculate churn rate for each TenureGroup
churn_analysis = data.groupby('TenureGroup')['Churn'].value_counts(normalize=True).unstack()['Yes'] * 100

# Display the churn rates
print(churn_analysis)
```

```python

# Plot churn rates
churn_analysis.plot(kind='bar', color='skyblue', figsize=(10, 6))

# Add labels and title
plt.title("Churn Rate by Tenure Group", fontsize=16)
plt.xlabel("Tenure Group", fontsize=12)
plt.ylabel("Churn Rate (%)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
